```python
###### Role Preference ######
dict_role={
    "Product Engineering / Development / Manufacturing)":["Product Launch/test opportunities"],
    "Make Investments":["Capital raise"],
    "Partnerships (Strategic / Joint Ventures / Business Development)":["Go-to-market Strategic Partners","Business/Joint Venture partners - Long term"],
    "Joint Ventures":["Business/Joint Venture partners - Long term"], 
    "Advisory support":["Expand business long-term into Asia","Legal and Regulatory Affairs advisory"], 
    "Offer Pilot tests":["Product Launch/test opportunities"],
    "Services":["Engineering (Technology / Datasets / Skilled labor / Software Development)","HR - Compiling local team","Office Services","Corporate Services (Incorporation, Re-structuring etc.)","Financial Services","Marketing and Digital Media Strategy (digital marketing, WeChat, Weibo, Social Media, SEO etc.)"],
    "Becoming Corporate Clients":["Making sales to corporate clients"],
    "Manufacturing":["XXXXX"], # does not exist in the form
    "Other":["XXXXX"],
}

##### Category group #######
dict_categorygroup={
    "Real Estate (Construction / Urban Development / Infrastructure etc.)":["Real Estate"],
    "Advertising":["XXX"],
    "Agriculture and Farming":["XXX"],
    "Healthcare / Biotechnology":["Biotechnology, Health Care"],
    "Manufacturing":["XXX"],
    "Transportation / Automotive":["XXX"],
    "Energy / Natural Resources / Sustainability":["XXX"],
    "Media and Entertainment":["XXX"],
}



###### Funding stage #########
dict_fundingstage={
    "No preference":["Pre-Seed","Seed","Series A","Series B", "Series C"],
    "Early stage":["Pre-Seed","Seed","Series A"],
    "Mid-term stage - Series A and Series B":["Pre-Seed","Series A","Series B"],
    "Late stage - post Series C round":["XXX","XXX"],
    "Publicly Listed companies":["XXX"],
    "Other":["XXX"]
}

########## Region #############
dict_region={
    "East Coast (New York City. NY. United States is the leading innovation hub in East cost)":["East Coast, US"],
    "West Coat (Silicon Valley CA. United States and surrounding regions)":["West Coast, US"],
    "Midwestern U.S":["Midwestern US"],
    "Southern U.S":["Southern US"],
    "Tel Aviv Israel":["XXX"],
    "Other":["XXX"], 
}

```

```python
dict_success={}
for i in range(0,len(client)):
    print(i)
    success_id=[]
    # role preferences
    roles=[]
    role_list=client['Preferred methods or traditionally adopted approaches to engage with startups?\n'][i].split(', ')
    for r in role_list:
        try:
            roles+=dict_role[r]
        except:
            print("Fail to look for this in the dict_role: "+r)
    roles=list(set(roles))
    # funding stages
    fundingstages=[]
    funding_list=client["Preferred Funding Stage:"][i].split(', ')
    for f in funding_list:
        try:
            fundingstages+=dict_fundingstage[f] # concatenate all the lists
        except:
            print("Fail to look for this in the dict_fundingstage: "+f)
    fundingstages=list(set(fundingstages))
    # regions
    regions=[]
    region_list=client["Which regions are you interested in expanding your innovation collaborations/sourcing efforts more?\n"][i].split(', ')
    for r in region_list:
        try:
            regions+=dict_region[r]
        except:
            print("Fail to look for this in the dict_region: "+r)
    regions=list(set(regions))
    # subcategories
    subcategories=[]
    subcate=client['sub-categories/innovation areas within {{answer_106901288}}important to the corporations long term innovation goals'][i]+', '+client['Industries/categories/innovation areas important to the corporations long term goals\n'][i]
    subcate_list=subcate.split(', ')
    
    for j in range(0,len(startup)):
        ################### Filtering Category Group ##################
        if(dict_categorygroup[client['Industry of your business?'][i]][0] != startup['Category Group'][j]):
            continue
        ################## Filtering Funding Stage ################
        if(startup['Stage '][j] not in fundingstages):
            continue
        ################## Filtering Region ######################
        if(startup['Region'][j] not in regions):
            continue
        ################## Filtering Subcategories #################
        condition1=False
        for x in startup['Categories'][j].split(', '):
            if(x in subcate_list):
                condition1=True
        if(condition1==False):
            continue
        ################### Filtering Role prefernce vs Needs ###################
        condition2=False
        for x in startup['Cross-border Needs'][j].split(', '):
            if(x in roles):
                condition2=True
        if(condition2==False):
            continue
        #### Sub Services #####
        services=client['What kind of Services can you provide / strategically offer to foreign startups?'][i].split(', ')
        condition3=True
        if("Services" in role_list):
            condition3=False
            for x in startup['Cross-border Needs'][j].split(', '):
                if(x in services):
                    condition3=True
                    #print("Success1")
            role_list_withoutservice=client['Preferred methods or traditionally adopted approaches to engage with startups?\n'][i].split(', ')
            role_list_withoutservice.remove('Services')
            roles_withoutservice=[]
            for r in role_list_withoutservice:
                try:
                    roles_withoutservice+=dict_role[r]
                except:
                    print("Fail to look for this in the dict_role: "+r)
                roles_withoutservice=list(set(roles_withoutservice))           
            for x in startup['Cross-border Needs'][j].split(', '):
                if(x in roles_withoutservice):
                    condition3=True
            #print("Success2")
        if(condition3==False):
            continue
            
        #################### Filtering Subcategories ##########
        success_id+=[j]
        print(" Matched client id: "+str(i)+" and startup id: "+str(j))
    dict_success[i]=success_id
        
```
